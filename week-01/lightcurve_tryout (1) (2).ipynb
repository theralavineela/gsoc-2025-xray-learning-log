```python
from stingray import Lightcurve
import numpy as np
import matplotlib.pyplot as plt
```

## üî¢ Simulate Time and Counts

```python
time = np.linspace(0, 100, 500)  # 500 points over 100 seconds
counts = np.random.poisson(lam=10, size=500)  # Poisson-distributed counts
```

## üìä Create and View Light Curve

```python
lc = Lightcurve(time, counts)
print("Time bin size:", lc.dt)
print("Total duration:", lc.tseg)
print("Mean count rate:", np.mean(lc.counts))
```

```python
lc.plot(title="Simulated Light Curve (Poisson Noise)")
```

## üåü Add a Synthetic Flare

```python
flare_counts = counts.copy()
flare_counts[230:270] += 30  # Add a flare around the middle
lc_flare = Lightcurve(time, flare_counts)
lc_flare.plot(title="Light Curve with Synthetic Flare")
```
